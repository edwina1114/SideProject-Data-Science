## Data Description

#### Data Set:  
[Spotify Tracks](https://www.kaggle.com/datasets/gauthamvijayaraj/spotify-tracks-dataset-updated-every-week/data)

#### Column Description:

- track_id: Spotify ID for the track.

- track_name: Name of the track.

- artist_name: The names of artists who performed the track, separated by commas if there are multiple artists.

- year: The release year of the track.

- popularity: A value between 0 and 100 indicating how popular a track is based on plays and recency.

- artwork_url: URL of the album or track's artwork.

- album_name: The album in which the track appears.

- acousticness: A confidence measure from 0.0 to 1.0 of whether the track is acoustic.

- danceability: A measure of how suitable a track is for dancing (0.0 = least danceable, 1.0 = most danceable).

- duration_ms: Track length in milliseconds.

- energy: A perceptual measure from 0.0 to 1.0 of intensity and activity.

- key: The musical key of the track, using standard pitch class notation (e.g., 0 = C, 1 = C♯/D♭).

- liveness: A measure of the likelihood that the track was recorded live (higher values indicate live performances).

- loudness: The overall loudness of the track in decibels (dB).

- mode: Indicates the modality of the track (1 = major, 0 = minor).

- speechiness: Measures the presence of spoken words in a track (closer to 1.0 indicates more speech-like content).

- tempo: The estimated tempo of the track in beats per minute (BPM).

- time_signature: The number of beats per measure, ranging from 3 to 7.

- valence: A measure from 0.0 to 1.0 indicating the track's musical positiveness (higher values are happier).

- track_url: The Spotify URL for the track.

- language: The language of the track (English, Tamil, Hindi, Telugu, Malayalam, Korean).

## Exploratory Data Analysis (EDA) 

## PART 1 - Which Language of Songs Became Popular Over the Years?

**1. Dominance of English Tracks in the Early Years**
   - **Exclusive Popularity**: In the early years, English tracks were the only language represented and had the highest popularity.
   - **Potential Dataset Bias**: The absence of non-English tracks could be due to dataset limitations.

**2. Sharp Rise of Hindi Music in 2007**
   - **Cultural Phenomena**: The spike in Hindi music's popularity in 2007 likely aligns with major Bollywood movie releases. Blockbuster films often feature soundtracks that resonate widely, driving their popularity.

**3. Decline in English Track Popularity**
   - **Increased Diversity**: As songs in other languages began appearing on Spotify, the dominance of English tracks started to decline.

**4. Rise of K-pop**
   - **Global Success Post-2010**: The rise of K-pop is evident after 2010. This surge is attributed to the iconic groups and viral hits.
   - **Cultural Export**: K-pop’s global success demonstrates the power of strategic marketing, visually captivating performances, and fan engagement.

**Conclusion:**
   - In recent years, K-pop has emerged as one of the most popular music genres globally. 

```python
popularity_year_language = sp_raw_data.groupby(['year', 'language'])['popularity'].mean().reset_index()
popularity_year_language
```

```python
# Interactive Plot
import plotly.express as px

fig = px.line(
    popularity_year_language,
    x='year',
    y='popularity',
    color='language',
    title='Popularity by Year and Language',
    labels={'popularity': 'Popularity', 'year': 'Year'},
)
fig.show()
```

## Part 2 - Sharp Rise of Hindi Music in 2007.

In 2007, Hindi songs achieved unprecedented popularity despite only three songs being in the dataset compared to 344 English songs. This success highlights the cultural power of Bollywood, as all three songs were tied to blockbuster films. Their immense popularity, driven by the success of these movies and the growing accessibility of online music platforms, marked a pivotal moment when Bollywood's influence expanded beyond India.



```python
sp_2007 = sp_raw_data[sp_raw_data['year'] == 2007]

fig = px.histogram(
    data_frame=sp_2007,
    x='language',
    color='language',
    title='Counts of Languages of Songs in 2007',
    labels={'language': 'Language', 'count': 'Counts of Songs'}
)
fig.update_layout(yaxis_title="Counts of Songs")
fig.show()
```

### Find 3 songs of Hindi.

```python
hindi_2007 = sp_2007[sp_2007['language'] == 'Hindi']
hindi_2007
```

### Comparison of Hindi Songs with Other Languages in 2007

**1. Distinctiveness of Hindi Music:**

Hindi music exhibits significant differences in its musical features compared to other languages, often ranking either the highest or the lowest across various attributes.

**2. Energetic and Upbeat Nature:**

* High danceability, tempo and valence:
Hindi music is characterized by its energetic, with fast-paced rhythms and an uplifting, joyful tone.

**3. Lower Emphasis on Vocals:**

* Low speechiness: The low speechiness indicates that Hindi songs typically have more melodic or lyrical content rather than spoken words or rap.

**4. Simpler Musical Structure:**

* Low key, liveness, and mode: Hindi music features a simpler key range and a lower emphasis on live elements, suggesting a preference for studio-produced tracks. The lower mode may indicate that Hindi music often leans toward more minor tonalities, despite its upbeat nature.

```python
''' The relationship between the average popularity of each language and 
the average values of other features. '''
numeric_columns_2007 = sp_2007.select_dtypes(include=['int64', 'float64']).columns
hindi_popularity = sp_2007.groupby('language')[numeric_columns_2007].mean().reset_index()
hindi_popularity
```

```python
# Correlation
cor_2007 = hindi_popularity.drop(['language', 'year'], axis=1).corr()
cor_2007
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(cor_2007, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Music Features Across Languages (2007)')
plt.show()
```

### Identifying High-Correlation Features for Cross-Language Investigation.

```python
hindi_features = ['danceability', 'key', 'liveness', 'mode', 'speechiness', 'tempo', 'time_signature', 'valence']
plt.figure(figsize=(12, 10))

for i, feature in enumerate(hindi_features):
    plt.subplot(3, 3, i+1)
    sns.scatterplot(x='language', y=feature, data=hindi_popularity, hue='language', legend=False)
    plt.title(f"Language vs {feature}")
plt.tight_layout()
plt.show()
```

## Part 3 - Rise of K-pop

2017開始K-POP成為最高人氣，所以先抓2017以後的資料，來看哪些歌曲元素與人氣有關，再來研究KPOP的音樂特色與其他語言不同處。

```python
songs_2017 = sp_raw_data[sp_raw_data.year>=2017]
songs_2017
```

### Comparison of K-POP with Other Languages after 2017

**1. K-pop has the shortest song duration.**

**2. K-POP emphasizes danceability：**
* Energy: K-pop generally have the highest energy, contributing to a dynamic and upbeat atmosphere that enhances its appeal in performances and music videos.

* Tempo: The tempo of K-pop tracks tends to be faster, creating a sense of urgency and excitement that drives the rhythm and mood of the music.

* Time Signature:  K-pop often features higher time signatures (ranging from 3 to 7 beats per measure), resulting in more complex and varied rhythmic structures, which is a defining feature of K-pop performances.

* Acousticness: K-pop tracks typically have lower acousticness, reflecting a greater reliance on electronic production, synthesized sounds, and studio effects.

**3. K-pop is strongly focused on live performance :**
* Liveness: K-pop has a notable emphasis on live performances, with artists regularly showcasing their talent in concerts, television shows, and other events. The "liveness" factor is one of the highest, often blending energetic stage presence with real-time vocal and dance performances, setting it apart from other music genres.

```python
numeric_columns_2017 = songs_2017.select_dtypes(include=['int64', 'float64']).columns
comp_songs_2017 = songs_2017.groupby('language')[numeric_columns_2017].mean().reset_index()
comp_songs_2017
```

```python
# Correlation
cor_songs_2017 = comp_songs_2017.drop(['language','year'], axis=1).corr()
cor_songs_2017
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(cor_songs_2017, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Music Features Across Languages (2017)')
plt.show()
```

```python
k_features = ['acousticness', 'energy', 'duration_ms', 'instrumentalness', 'liveness', 'loudness', 'tempo', 'time_signature']
plt.figure(figsize=(12, 10))

for i, feature in enumerate(k_features):
    plt.subplot(3, 3, i+1)
    sns.scatterplot(x='language', y=feature, data=comp_songs_2017, hue='language', legend=False)
    plt.title(f"Language vs {feature}")
    plt.xticks(rotation=45)
    
plt.tight_layout()
plt.show()
```
